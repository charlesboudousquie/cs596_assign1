__Question 1, Part 1: Check for missing data and data duplication__

```python
# Check for missing, duplicates, and negatives

# Note: For code written by Charles Boudousquie, "Geeksforgeeks.com", "pandas.pydata.org", 
# and "claude ai sonnet version 4" were used as a reference to find useful pandas 
# functions like ".any(axis=1)' for question 1, but the actual code was written by
# me (Charles Boudousquie).

# Note: You can use .to_string() on certain objects
# when calling print to force pandas to display each value. 
# Example: print(missing_rows.to_string())


# First get a (boolean) dataframe showing which values are missing
missing_values_mask = iris_df.isnull()
# Then get a pandas series(a list of booleans, 1 per row) showing which 
# rows have at least 1 missing value.
# Axis 0 would be columns names, but we want the actual rows which would be
# axis 1.
missing_rows_mask = missing_values_mask.any(axis=1)

# Now we get the actual missing rows(aka records),
# Note: missing_rows is NOT a copy, it is a view into the data
# inside iris_df
missing_rows = iris_df[missing_rows_mask]
# Display records with missing values.
# As we can see only the sepal width column has missing values.
display(missing_rows)

# average sepal width is 3.0505301438798096
sepal_column = iris_df.loc[:, 'sepal width (cm)']
sepal_width_avg = sepal_column.mean()
sepal_width_median = sepal_column.median()
print('sepal width average: ', sepal_width_avg)
# Sepal width average and median are close
# Sepal median: 3.033138148372734
print('sepal width median: ', sepal_width_median)

```
